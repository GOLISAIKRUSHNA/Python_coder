```python
#power of 2 using bitwise shift operator 1<<i


n=int(input())

i=0
res=0
while res<n:
    res=1<<i  #2^0=1,10,11,100,101
        #leftshift operator shift by 1

    if res==n:
        print("power of 2",i)
        break
    i=i+1
else:
    print("not a power of 2")

# while i<n:
#     if(2**i==n):
#         print("power",i)
#         break
#     i=i+1
# else:
#     print("not a power")
   


















```
